```python
# y = [df_exp102['      train/box_loss'], df_exp112['      train/box_loss']]
# x = [df_exp102['               epoch'], df_exp112['               epoch']]

# a = df_exp102.plot('               epoch','      train/box_loss', label = '102')
# b = df_exp112.plot('               epoch','      train/box_loss', label = '112')

plt.figure(figsize=(600,600))

figure, axis = plt.subplots(2, 5, figsize = (20,10))
# axis = plt.subplots(2, 5, figsize = (20,10))


axis[0, 0].plot(df_exp102['               epoch'], df_exp102['      train/box_loss'], color='r', label='YOLO w/0 finetune')
axis[0, 0].plot(df_exp112['               epoch'], df_exp112['      train/box_loss'], color='g', label='YOLO w/ finetune')
axis[0, 0].set_title("train/box_loss", fontsize = 15)

axis[0, 1].plot(df_exp102['               epoch'], df_exp102['      train/obj_loss'], color='r', label='YOLO w/0 finetune')
axis[0, 1].plot(df_exp112['               epoch'], df_exp112['      train/obj_loss'], color='g', label='YOLO w/ finetune')
axis[0, 1].set_title("train/obj_loss", fontsize = 15)

axis[0, 2].plot(df_exp102['               epoch'], df_exp102['      train/cls_loss'], color='r', label='YOLO w/0 finetune')
axis[0, 2].plot(df_exp112['               epoch'], df_exp112['      train/cls_loss'], color='g', label='YOLO w/ finetune')
axis[0, 2].set_title("train/cls_loss", fontsize = 15)

axis[0, 3].plot(df_exp102['               epoch'], df_exp102['   metrics/precision'], color='r', label='YOLO w/0 finetune')
axis[0, 3].plot(df_exp112['               epoch'], df_exp112['   metrics/precision'], color='g', label='YOLO w/ finetune')
axis[0, 3].set_title("metrics/precision", fontsize = 15)

axis[0, 4].plot(df_exp102['               epoch'], df_exp102['      metrics/recall'], color='r', label='YOLO w/0 finetune')
axis[0, 4].plot(df_exp112['               epoch'], df_exp112['      metrics/recall'], color='g', label='YOLO w/ finetune')
axis[0, 4].set_title("metrics/recall", fontsize = 15)

axis[1, 0].plot(df_exp102['               epoch'], df_exp102['        val/box_loss'], color='r', label='YOLO w/0 finetune')
axis[1, 0].plot(df_exp112['               epoch'], df_exp112['        val/box_loss'], color='g', label='YOLO w/ finetune')
axis[1, 0].set_title("val/box_loss", fontsize = 15)

axis[1, 1].plot(df_exp102['               epoch'], df_exp102['        val/obj_loss'], color='r', label='YOLO w/0 finetune')
axis[1, 1].plot(df_exp112['               epoch'], df_exp112['        val/obj_loss'], color='g', label='YOLO w/ finetune')
axis[1, 1].set_title("val/obj_loss", fontsize = 15)

axis[1, 2].plot(df_exp102['               epoch'], df_exp102['        val/cls_loss'], color='r', label='YOLO w/0 finetune')
axis[1, 2].plot(df_exp112['               epoch'], df_exp112['        val/cls_loss'], color='g', label='YOLO w/ finetune')
axis[1, 2].set_title("val/cls_loss", fontsize = 15)

axis[1, 3].plot(df_exp102['               epoch'], df_exp102['     metrics/mAP_0.5'], color='r', label='YOLO w/0 finetune')
axis[1, 3].plot(df_exp112['               epoch'], df_exp112['     metrics/mAP_0.5'], color='g', label='YOLO w/ finetune')
axis[1, 3].set_title("metrics/mAP_0.5", fontsize = 15)

axis[1, 4].plot(df_exp102['               epoch'], df_exp102['metrics/mAP_0.5:0.95'], color='r', label='YOLO w/0 finetune')
axis[1, 4].plot(df_exp112['               epoch'], df_exp112['metrics/mAP_0.5:0.95'], color='g', label='YOLO w/ finetune')
axis[1, 4].set_title("metrics/mAP_0.5:0.95", fontsize = 15)


plt.legend(bbox_to_anchor=(1.5, 1.5), loc='upper left')
figure.suptitle('YOLO baseline models comparsion', fontsize = 20)

plt.legend(prop={"size":12})
plt.show()
```
