```python
#converts DCIM files into jpegs. NOT WORKING PROPERLY (issue: most panoramic radiographs save as a white screen instead of the radiograph itself)
import pydicom as dicom
import os
import cv2
#from PIL import Image
# make it True if you want in PNG format
PNG = False
# Specify the .dcm folder path
folder_path = "newXray\\ps"
# Specify the output jpg/png folder path
jpg_folder_path = "newXray\\psTojpg"
images_path = os.listdir(folder_path)
for n, image in enumerate(images_path):
    ds = dicom.dcmread(os.path.join(folder_path, image))
    pixel_array_numpy = ds.pixel_array
    if PNG == False:
        image = image.replace('.DCM', '.jpg')
        image = image.replace('.dcm', '.jpg')
    else:
        image = image.replace('.dcm', '.png')
    print(os.path.join(jpg_folder_path, image))
    cv2.imwrite(os.path.join(jpg_folder_path, image), pixel_array_numpy)
    if n % 50 == 0:
        print('{} image converted'.format(n))
```
