```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# mat = np.array([[1, 3, 2], [5, 6, 4]])
cf_matrix = np.array([[116,   0, 279], [0,   0,   0], [78,  26,   0]])
```

```python
group_counts = ["{0:0.0f}".format(value) for value in
                cf_matrix.flatten()]

group_percentages = ["{0:.2%}".format(value) for value in
                     cf_matrix.flatten()/np.sum(cf_matrix)]

labels = [f"{v1}\n{v2}" for v1, v2 in
          zip(group_counts,group_percentages)]

labels = np.asarray(labels).reshape(3,3)

ax = sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues', annot_kws={"size":14})

# ax.set_title('Confusion Matrix\n\n');
ax.set_xlabel('\nPredicted Values', fontsize=14)
ax.set_ylabel('Actual Values', fontsize=14);

## Ticket labels - List must be in alphabetical order
# ax.xaxis.set_ticklabels(['False','True'])
# ax.yaxis.set_ticklabels(['False','True'])

## Display the visualization of the Confusion Matrix.
plt.show()
```
